```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()

```

```python
plt.figure(figsize=(10, 8))
plt.scatter(df.hours,df.grade,alpha=0.5,color="purple")
plt.title("Scatter Plot of hours and grades")
plt.xlabel("hours")
plt.ylabel("grades")
plt.show()
```

Yes there is a linear relationship. More the hours more is the grade
