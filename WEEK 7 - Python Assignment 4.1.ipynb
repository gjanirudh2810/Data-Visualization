```python
import pandas as pd
import numpy as np
Location = "datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
# Create the bin dividers
bins = [0, 60, 70, 80, 90, 100]
# Create names for the four groups
group_names = ['F', 'D', 'C', 'B', 'A']
```

```python
df['lettergrade'] = pd.cut(df['grade'], bins,
        labels=group_names)
df.head()
```

```python
result=[]
for row in df['grade']:
    if row>80:
        result.append("pass")
    else:
        result.append("fail")
df['result']=result
```
